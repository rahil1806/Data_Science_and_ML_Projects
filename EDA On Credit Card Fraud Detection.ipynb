```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('creditcard.csv')
df.shape
```

```python
#plot distribution
sns.countplot(x='Class', data=df);
plt.title('Class Distribution');
```

```python
#calculating the fraud rate
df['Class'].value_counts(normalize=True)
```

```python
# Amount Distribution
sns.histplot(df['Amount'], bins=50, kde=True, color='red')
plt.title("Amount Distribution")
plt.show()
```

```python
# Time Distribution
sns.histplot(df['Time'], bins=50, kde=True, color='green')
plt.title("Time Distribution")
plt.show()
```

```python
# Log-Transform Amount
df['LogAmount'] = np.log1p(df['Amount'])
sns.boxplot(x=df['LogAmount'])
plt.title("Boxplot of Log-Transformed Amount")
plt.show()
```

```python
#Correlation Analysis
corr = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr[['Class']].sort_values(by='Class', ascending=False),
            annot=True, cmap='coolwarm')
```

```python
#Comparision
fraud = df[df['Class']==1]
nonfraud = df[df['Class']==0]
df.groupby('Class')['Amount'].mean().plot(kind='bar', color=['blue', 'red'])
plt.title("Average Transaction Amount by Class")
plt.xticks(ticks=[0,1], labels=["Non-Fraud", "Fraud"], rotation=0)
plt.ylabel("Average Amount")
plt.show()


```

```python
#Time-based Analysis
df['Hour'] = (df.Time // 3600) % 24
sns.countplot(x='Hour', hue='Class', data=df);
```

```python
# Z-score for detecting Amount outliers
from scipy.stats import zscore
df['Amount_zscore'] = zscore(df['Amount'])
outliers = df[np.abs(df['Amount_zscore']) > 3]
print("Outliers in Amount:", len(outliers))
```

```python
# Scatter plot to show fraud anomalies
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, alpha=0.6, palette={0:'blue', 1:'red'})
plt.title("Time vs Amount (Fraud in Red)")
plt.show()
```
